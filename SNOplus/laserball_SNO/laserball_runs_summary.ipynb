```python
# Get the histogram for a given number of file
def getTimeHisto(file_nr, data):
    if data.has_key(file_nr):
        print 'Data already has that key! Exiting ...'
    else:
        print 'Adding key ', file_nr
    
    data[file_nr] = {}

    tfile = ROOT.TFile(os.path.join(data_dir, file_list[file_nr]))
    
    data[file_nr]['spos'] = (tfile.Get('source_pos'))
    data[file_nr]['lamb'] = np.array(tfile.Get('lambda'))
    data[file_nr]['cast'] = np.array(tfile.Get('cast'))
    data[file_nr]['version'] = np.array(tfile.Get('version'))
    
    # Getting the info out of this one file
    caltime = tfile.Get('fRchTime')
    
    xaxis = caltime.GetXaxis()
    yaxis = caltime.GetYaxis()
    nbinsx = caltime.GetNbinsX()
    nbinsy = caltime.GetNbinsY()
    
    xedges = np.zeros(nbinsx+1)
    yedges = np.zeros(nbinsy+1)
    
    out_histo = np.zeros([nbinsx, nbinsy])
    for xbin in range(nbinsx+1):
        xedges[xbin]    = xaxis.GetBinLowEdge(xbin)
        
    for ybin in range(nbinsy+1):
        yedges[ybin]    = yaxis.GetBinLowEdge(ybin)
        
    for xbin in range(nbinsx):
        for ybin in range(nbinsy):
            out_histo[xbin, ybin] = caltime.GetBinContent(xbin, ybin)
    data[file_nr]['toa'] = out_histo
    data[file_nr]['xedges'] = xedges
    data[file_nr]['yedges'] = yedges
```
