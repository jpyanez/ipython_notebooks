```python
qpdt_prob_scale = 0.0143
qpdtqmax = 250
```

```python
tl208_type = 'tb6_mc_tl208'
indir = basedir +'/'+tl208_type+'_npy'
tl208={}
for one_var in var_names:
    tl208[one_var] = np.load(os.path.join(indir, one_var+'.npy'))
tl208_tcp = 1-(1-tl208['qpdtProbability'])**tl208['qpdtNhitsEarly']
tl208_sitr = (0.74-tl208['itr'])*np.sqrt(tl208['nhitsCleaned'])/0.43
```

```python
ecut = 3.25
```

```python
tl208_cuts = ((tl208['energy']>ecut) *
       #(all_pmt['energy']<4.)*
       (tl208['itr']>0.55)*
       (tl208['beta14']<0.95)*
       (tl208['inTimeHits100']>=23)*
       (tl208['beta14']>-0.12)*
       (tl208['posr']<5500.))
print np.sum(tl208_cuts)
#scale = len(pmt_cuts)*1./np.sum(pmt_cuts)
#print scale

poisson_cut = (stats.poisson.pmf(tl208['qpdtNhitsEarly'], tl208['nhitsCleaned']*qpdt_prob_scale) > 0.002)
prob_cut = ((1-(1-tl208['qpdtProbability'])**tl208['qpdtNhitsEarly']) > 0.01)
nhit_cut = (tl208['qpdtNhitsEarly']==0)
qmax_cut = (tl208['qpdtQMax']<qpdtqmax)

# Efficiency of each cut
print 'Poisson', 1.*np.sum(poisson_cut*tl208_cuts)/np.sum(tl208_cuts)
print 'Probability', 1.*np.sum(prob_cut*tl208_cuts)/np.sum(tl208_cuts)
print 'QMax', 1.*np.sum(qmax_cut*tl208_cuts)/np.sum(tl208_cuts)
print 'Combined', 1.*np.sum(prob_cut*tl208_cuts*poisson_cut)/np.sum(tl208_cuts)
print 'Nhit', 1.*np.sum(nhit_cut*tl208_cuts)/np.sum(tl208_cuts)

#tlqpdtcut = (poisson_cut*prob_cut)+nhit_cut
tlqpdtcut = (qmax_cut+(poisson_cut*prob_cut))+nhit_cut

#tlqpdtcut = ((stats.poisson.pmf(tl208['qpdtNhitsEarly'], tl208['nhitsCleaned']*qpdt_prob_scale) > 0.002)*
#           (((1-(1-tl208['qpdtProbability'])**tl208['qpdtNhitsEarly']) > 0.01) + 
#           (tl208['qpdtNhitsEarly']==0)))

#qpdtcut = all_pmt_tcp > 0.01
print '\nEvents', np.sum(tl208_cuts*~tlqpdtcut)
relerror = np.sqrt(np.sum(tl208_cuts*~tlqpdtcut))/np.sum(tl208_cuts*~tlqpdtcut)
rejected = 100*np.sum(tl208_cuts*~tlqpdtcut)*1./np.sum(tl208_cuts)
print 'Rejected fraction', rejected , '+/-', rejected*relerror
scale = np.sum(tl208_cuts)*1./np.sum(tlqpdtcut*tl208_cuts)
print scale
```

```python
pmt_cuts = ((all_pmt['energy']>ecut) *
       #(all_pmt['energy']<3.5)*
       (all_pmt['beta14']<0.95)*
       (all_pmt['beta14']>-0.12)*
       #(all_pmt['inTimeHits100']>=23)*
       #(all_pmt['itr']>0.55)*
       (all_pmt['posr']<5500.))
print np.sum(pmt_cuts)
#scale = len(pmt_cuts)*1./np.sum(pmt_cuts)
#print scale

poisson_cut = (stats.poisson.pmf(all_pmt['qpdtNhitsEarly'], all_pmt['nhitsCleaned']*qpdt_prob_scale) > 0.002)
prob_cut = ((1-(1-all_pmt['qpdtProbability'])**all_pmt['qpdtNhitsEarly']) > 0.01)
nhit_cut = (all_pmt['qpdtNhitsEarly']==0)
qmax_cut = (all_pmt['qpdtQMax']<qpdtqmax)

# Efficiency of each cut
print 'Poisson', 1.*np.sum(poisson_cut*pmt_cuts)/np.sum(pmt_cuts)
print 'Probability', 1.*np.sum(prob_cut*pmt_cuts)/np.sum(pmt_cuts)
print 'QMax', 1.*np.sum(qmax_cut*pmt_cuts)/np.sum(pmt_cuts)
print 'Combined', 1.*np.sum(prob_cut*pmt_cuts*poisson_cut)/np.sum(pmt_cuts)
print 'Nhit', 1.*np.sum(nhit_cut*pmt_cuts)/np.sum(pmt_cuts)

#qpdtcut = (poisson_cut*prob_cut)+nhit_cut
qpdtcut = (qmax_cut+(poisson_cut*prob_cut))+nhit_cut


#qpdtcut = ((stats.poisson(all_pmt['qpdtNhitsEarly'], 
#             all_pmt['nhitsCleaned']*qpdt_prob_scale) > 0.002)*
#           (((1-(1-all_pmt['qpdtProbability'])**all_pmt['qpdtNhitsEarly']) > 0.01) + 
#           (all_pmt['qpdtNhitsEarly']==0)))
           



#qpdtcut = all_pmt_tcp > 0.01
print 'Events', np.sum(pmt_cuts*~qpdtcut)
relerror = np.sqrt(np.sum(pmt_cuts*~qpdtcut))/np.sum(pmt_cuts*~qpdtcut)
rejected = 100*np.sum(pmt_cuts*~qpdtcut)*1./np.sum(pmt_cuts)
print 'Rejected fraction', rejected , '+/-', rejected*relerror
scale = np.sum(pmt_cuts)*1./np.sum(qpdtcut*pmt_cuts)
print scale
```

```python
nbins = 22
xenergy = np.linspace(3.25, 4, nbins)
xcenergy = (xenergy[1:] + xenergy[:-1])/2.
xbeta14 = np.linspace(-0.12,0.95, nbins)
xradius = np.linspace(0, .8, nbins)
rav = 6000.
```

```python
data_type = 'tb1_data'
indir = basedir +'/'+data_type+'_npy'
data_tb6={}
var_names =['qpdtProbability',
            'qpdtNhitsEarly',
            'nhitsCleaned',
            'itr',
            'energy',
            'posz',
            'qpdtQMax',
            'beta14',
            'dcApplied',
            'dcFlagged',
            'posr']
for one_var in var_names:
    data_tb6[one_var] = np.load(os.path.join(indir, one_var+'.npy'))
data_tb6_tcp = 1-(1-data_tb6['qpdtProbability'])**data_tb6['qpdtNhitsEarly']
data_tb6_sitr = (0.74-data_tb6['itr'])*np.sqrt(data_tb6['nhitsCleaned'])/0.43
```

```python
mask_cut = np.zeros_like(data_tb6['energy'], dtype=bool)
for i in range(mask_cut.size):
    mask_cut[i] = (((int(data_tb6['dcApplied'][i]) & 0x10000017FFE) & int(data_tb6['dcFlagged'][i])) == 
                   (int(data_tb6['dcApplied'][i]) & 0x10000017FFE))
print np.sum(mask_cut)*1./mask_cut.size
```

```python
# Reproducing figure 16.13 - all for beta14 (light water region)
```

```python
lw_pmt = ((all_pmt['energy']>3.25) *
       (all_pmt['energy']<4.)*
       (all_pmt['beta14']<0.95)*
          #(all_pmt['posz']<0)*
       (all_pmt['nhitsCleaned'] > 15.)*
       (all_pmt['beta14']>-0.12)*
       (all_pmt['posr']>6100.))
print np.sum(lw_pmt)

lw_data = ((data_tb6['energy']>3.25) *
           mask_cut*
       (data_tb6['energy']<4.)*
           #(data_tb6['posz']<0)*
       (data_tb6['nhitsCleaned'] > 15.)*
       (data_tb6['beta14']<0.95)*
       (data_tb6['beta14']>-0.12)*
       (data_tb6['posr']>6100.))
print np.sum(lw_data)
#scale = len(pmt_cuts)*1./np.sum(pmt_cuts)
#print scale


qpdt_pmt = ((stats.poisson.pmf(all_pmt['qpdtNhitsEarly'], 
             all_pmt['nhitsCleaned']*qpdt_prob_scale) > 0.002)*
           (((1-(1-all_pmt['qpdtProbability'])**all_pmt['qpdtNhitsEarly']) > 0.01) + 
           (all_pmt['qpdtNhitsEarly']==0)) +
           (all_pmt['qpdtQMax'] < qpdtqmax))
qpdt_data = ((stats.poisson.pmf(data_tb6['qpdtNhitsEarly'], 
             data_tb6['nhitsCleaned']*qpdt_prob_scale) > 0.002)*
           (((1-(1-data_tb6['qpdtProbability'])**data_tb6['qpdtNhitsEarly']) > 0.01) + 
           (data_tb6['qpdtNhitsEarly']==0))+
            (data_tb6['qpdtQMax'] <qpdtqmax))
```

```python
fig = plt.figure(figsize=(22,4))
xbeta14 = np.linspace(-0.12,0.95, 41)

# Data passing QPDT cut
bdata_qpdt, x = np.histogram(data_tb6['beta14'][lw_data*qpdt_data], xbeta14)
bmc_qpdt, x = np.histogram(all_pmt['beta14'][lw_pmt*qpdt_pmt], xbeta14)

# All data
bdata_all, x = np.histogram(data_tb6['beta14'][lw_data], xbeta14)
bmc_all, x = np.histogram(all_pmt['beta14'][lw_pmt], xbeta14)


ax1 = fig.add_subplot(131)
jplot.errorMark(x, bdata_qpdt, color = 'k', label='Data wQPDT')
scale = bdata_qpdt.sum()*1./bmc_qpdt.sum()
print bmc_qpdt.sum()*scale, bdata_qpdt.sum()

relerror = np.sqrt(bmc_qpdt)/bmc_qpdt
jplot.errorMark(x, scale*bmc_qpdt, rel_error=relerror*scale, color='red', label='PMT MC wQPDT' )
plt.legend(loc=0)
plt.xlabel('beta14')

ax2 = fig.add_subplot(132)
scale = bdata_all.sum()*1./bdata_qpdt.sum()
print bdata_qpdt.sum()*scale, bdata_all.sum()
relerror = np.sqrt(bdata_qpdt)/bdata_qpdt
jplot.errorMark(x, scale*bdata_qpdt, rel_error=relerror*scale, color='black' , label='Data wQPDT')
jplot.errorMark(x, bdata_all, color = 'green', label='Data full')
plt.legend(loc=0)
plt.xlabel('beta14')

ax3 = fig.add_subplot(133)
jplot.errorMark(x, bmc_all, color = 'blue', label='MC full')
scale = bmc_all.sum()*1./bmc_qpdt.sum()
print bmc_qpdt.sum()*scale, bmc_all.sum()
relerror = np.sqrt(bmc_qpdt)/bmc_qpdt
jplot.errorMark(x, scale*bmc_qpdt, rel_error=relerror*scale, color='red' , label='PMT MC wQPDT')
plt.legend(loc=0)
plt.xlabel('beta14')
fig.savefig('/home/jpyanez/snoplus/analysis/water_analysis/qpdt_figures/Fig_16_13_beta14_data_mc.png', dpi=200)
```
