```python
basedir = '/home/jpyanez/scratch/snoplus/precuts'
```

```python
var_names =['qpdtProbability',
            'qpdtNhitsEarly',
            'nhitsCleaned',
            'itr',
            #'dcApplied',
            #'dcFlagged',
            'energy',
            'beta14',
            'posr']
```

```python
# Loading all the data
```

```python
#all_pmt_type = 'tb3_mc_tl208pmt'
all_pmt_type = 'all_pmt'
indir = basedir +'/'+all_pmt_type+'_npy'
all_pmt={}
for one_var in var_names:
    all_pmt[one_var] = np.load(os.path.join(indir, one_var+'.npy'))
```

```python
bi214_type = 'tb6_mc_bi214'
indir = basedir +'/'+bi214_type+'_npy'
bi214={}
for one_var in var_names:
    bi214[one_var] = np.load(os.path.join(indir, one_var+'.npy'))
```

```python
data_type = 'tb6_data'
var_names += ['dcFlagged', 'dcApplied']
indir = basedir +'/'+data_type+'_npy'
data_tb6={}
for one_var in var_names:
    data_tb6[one_var] = np.load(os.path.join(indir, one_var+'.npy'))
```

```python
# Cuts
```

```python
mine = 3.
maxe = 4.
maxr = 5500.

bi214_cuts = ((bi214['energy']>mine) *
       (bi214['energy']<maxe)*
       #(tl208['itr']>0.55)*
       (bi214['beta14']<0.95)*
       (bi214['beta14']>-0.12)*
       (bi214['posr']<maxr))
print np.sum(bi214_cuts)
pmt_cuts = ((all_pmt['energy']>mine) *
       (all_pmt['energy']<maxe)*
       (all_pmt['beta14']<0.95)*
       (all_pmt['beta14']>-0.12)*
       #(all_pmt['itr']>0.55)*
       (all_pmt['posr']<maxr))
print np.sum(pmt_cuts)
```

```python
# These are the new variables
hits_sitr = 'nhitsCleaned'
#hits_sitr = 'qpdtNhitsEarly'

data_tb6_tcp = 1-(1-data_tb6['qpdtProbability'])**data_tb6['qpdtNhitsEarly']
data_tb6_sitr = (0.74-data_tb6['itr'])*np.sqrt(data_tb6[hits_sitr])/0.43

all_pmt_tcp = 1-(1-all_pmt['qpdtProbability'])**all_pmt['qpdtNhitsEarly']
all_pmt_sitr = (0.74-all_pmt['itr'])*np.sqrt(all_pmt[hits_sitr])/0.43

bi214_tcp = 1-(1-bi214['qpdtProbability'])**bi214['qpdtNhitsEarly']
bi214_sitr = (0.74-bi214['itr'])*np.sqrt(bi214[hits_sitr])/0.43
```

```python
sitr_cut = 2.7
qpdt_cut = 0.01
```

```python


plt.plot(bi214_sitr[bi214_cuts], 
         bi214_tcp[bi214_cuts], '.', markersize=0.1, color='green',
        label='Bi214 MC')
plt.plot(all_pmt_sitr[pmt_cuts], 
         all_pmt_tcp[pmt_cuts], '.', markersize=.3, color ='k',
        label='All PMT MC')
plt.axvline(sitr_cut, 1E-4, 1, color='red')
plt.axhline(qpdt_cut, -3, 1, color='red')
plt.xlim(-2.5, 8)
plt.ylim(1E-4, 1)
plt.yscale('log')
plt.xlabel('Scaled ITR')
plt.ylabel('QPDT Charge probability')
plt.legend(loc=0)
```

```python
# Validation 1) Scaled ITR - 2) QPDT

# Bi214
pass_pass = (bi214_sitr[bi214_cuts]<sitr_cut)*(bi214_tcp[bi214_cuts]>qpdt_cut)
pass_fail = (bi214_sitr[bi214_cuts]<sitr_cut)*(bi214_tcp[bi214_cuts]<qpdt_cut)
fail_pass = (bi214_sitr[bi214_cuts]>sitr_cut)*(bi214_tcp[bi214_cuts]>qpdt_cut)
fail_fail = (bi214_sitr[bi214_cuts]>sitr_cut)*(bi214_tcp[bi214_cuts]<qpdt_cut)
ppsum = np.sum(pass_pass)
pfsum = np.sum(pass_fail)
fpsum = np.sum(fail_pass)
ffsum = np.sum(fail_fail)

pperr = np.sqrt(ppsum)/ppsum
pferr = np.sqrt(pfsum)/pfsum
fperr = np.sqrt(fpsum)/fpsum
fferr = np.sqrt(ffsum)/ffsum

print '\nBoxes (signal last)'
print pfsum, '+/-', pferr*pfsum
print fpsum, '+/-', fperr*fpsum
print ffsum, '+/-', fferr*ffsum
print ppsum, '+/-', pperr*ppsum

print '\nSignal box estimate'
estimate = pfsum*fpsum*1./ffsum
print estimate, '+/-', np.sqrt(pferr**2 + fperr**2+fferr**2)*estimate




ppsum = np.sum((all_pmt_sitr[pmt_cuts]<sitr_cut)*(all_pmt_tcp[pmt_cuts]>qpdt_cut))
pfsum = np.sum((all_pmt_sitr[pmt_cuts]<sitr_cut)*(all_pmt_tcp[pmt_cuts]<qpdt_cut))
fpsum = np.sum((all_pmt_sitr[pmt_cuts]>sitr_cut)*(all_pmt_tcp[pmt_cuts]>qpdt_cut))
ffsum = np.sum((all_pmt_sitr[pmt_cuts]>sitr_cut)*(all_pmt_tcp[pmt_cuts]<qpdt_cut))

pperr = np.sqrt(ppsum)/ppsum
pferr = np.sqrt(pfsum)/pfsum
fperr = np.sqrt(fpsum)/fpsum
fferr = np.sqrt(ffsum)/ffsum

print '\nBoxes (signal last)'
print pfsum, '+/-', pferr*pfsum
print fpsum, '+/-', fperr*fpsum
print ffsum, '+/-', fferr*ffsum
print ppsum, '+/-', pperr*ppsum

print '\nSignal box estimate'
estimate = pfsum*fpsum*1./ffsum
print estimate, '+/-', np.sqrt(pferr**2 + fperr**2+fferr**2)*estimate

```

```python

add_sitr = 0.1
rem_qpdt = 0.001

nsteps = 5

box = np.zeros(nsteps)
boxerr = np.zeros(nsteps)
est = np.zeros(nsteps)
esterr = np.zeros(nsteps)

for i in range(0, nsteps):

    ppsum = np.sum((all_pmt_sitr[pmt_cuts]<(sitr_cut+i*add_sitr))*(all_pmt_tcp[pmt_cuts]>(qpdt_cut-i*rem_qpdt)))
    pfsum = np.sum((all_pmt_sitr[pmt_cuts]<(sitr_cut+i*add_sitr))*(all_pmt_tcp[pmt_cuts]<(qpdt_cut-i*rem_qpdt)))
    fpsum = np.sum((all_pmt_sitr[pmt_cuts]>(sitr_cut+i*add_sitr))*(all_pmt_tcp[pmt_cuts]>(qpdt_cut-i*rem_qpdt)))
    ffsum = np.sum((all_pmt_sitr[pmt_cuts]>(sitr_cut+i*add_sitr))*(all_pmt_tcp[pmt_cuts]<(qpdt_cut-i*rem_qpdt)))

    pperr = np.sqrt(ppsum)/ppsum
    pferr = np.sqrt(pfsum)/pfsum
    fperr = np.sqrt(fpsum)/fpsum
    fferr = np.sqrt(ffsum)/ffsum

    est[i] = pfsum*fpsum*1./ffsum
    esterr[i] = np.sqrt(pferr**2 + fperr**2+fferr**2)*est[i]
    
    box[i] = ppsum
    boxerr[i] = ppsum*pperr

xax = range(0,nsteps)
fig = plt.figure()
plt.errorbar(xax, box, boxerr, fmt='x', label='Observed')
plt.errorbar(xax, est, esterr, fmt='o',label='Predicted')

print 'Increase', (box[-1]-box[0])/box[0]
plt.xlabel('Cut relaxation iteration')
plt.ylabel('PMT MC events')
plt.ylim(0, box.max()*1.8)
plt.xlim(xax[0]-0.5, xax[-1]+0.5)
plt.legend(loc=0)
```
