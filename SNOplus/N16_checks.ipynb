```python
def getRatioError(a, sigma_a, b, sigma_b):
    f = a/b
    return f*np.sqrt((sigma_a/a)**2 + (sigma_b/b)**2)
    
```

```python
plt.figure(figsize=(10,8))

# Ratio of RAT MC and SNO data
ratio = nhit_RATMC/nhit_SNO
ratio /= ratio[6]
plt.plot(position_x, ratio, 'xb', label='Ratio nhit RAT/SNO')
ratio_error = getRatioError(nhit_RATMC, Error_RATMC, nhit_SNO, Error_SNO)
plt.errorbar(position_x, ratio, yerr= ratio_error, fmt="none")

# Ratio of RAT MC and SNO data - prompt light only

ratio = nhit_promptRATMC/nhit_promptSNO
ratio /= ratio[6]
plt.plot(position_x, ratio, 'xg', label='Ratio prompt nhit RAT/SNO')
ratio_error = getRatioError(nhit_promptRATMC, Error_promptRATMC, nhit_promptSNO, Error_promptSNO)
plt.errorbar(position_x, ratio, yerr= ratio_error, fmt="none")

# Ratio of RAT MC and SNO data - using GDISC in RAT
ratio = nhit_GrayDiscRATMC/nhit_SNO
ratio /= ratio[6]
plt.plot(position_x, ratio, 'xr', label='Ratio nhit GDiscRAT/SNO')
ratio_error = getRatioError(nhit_GrayDiscRATMC, Error_GrayDiscRATMC, nhit_SNO, Error_SNO)
plt.errorbar(position_x, ratio, yerr= ratio_error, fmt="none")


plt.ylim([0.94, 1.06])
plt.xlabel('position (mm)')
plt.legend(loc=0)
```
