```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Water values
Z = 8 # Although it could be from 6-4 depending on whether you use equivalent due to shielding
re = 2.8*10**(-15) # m
Ek = 5.E6 # eV
me = 510.E3/c**2 # eV
c  = 3E8
beta = 1. # Approximation, should be v/c
ne = 1. # Electron density
e  = 1. # electron charge
```

```python
wp = (4*np.pi*ne*e**2/me)**(0.5) # Units of c
#wp = 90E6
```

```python
def ekpart(Ek):
    eratio = (Ek/(me*c**2))
    beta = np.sqrt(1-(1./(eratio+1)**2))
    numerator = eratio**2*beta*np.log(137.*Z**(-1/3.))
    denominator = ((1+eratio)**2-1)**2
    return numerator/denominator
```

```python
energies = np.linspace(500000,10E6, 101)
```

```python
plt.plot(energies/1E6, ekpart(energies)/ekpart(1E6))
plt.ylabel('q / q(1 MeV)')
plt.xlabel('E [MeV]')
```

```python
plt.plot(energies/1E6, ekpart(energies))
plt.yscale('log')
plt.ylabel('q')
plt.xlabel('E [MeV]')
```
