```python
indir = '/home/jp/projects/snoplus/rundir/greydisc_validation/onepmt_gdisc_model'
infile_name = ''
infile = open(os.path.join(indir, infile_name))
```

```python
# Checking the TDelay line
for line in infile:
    # Find the string of text
    if 'TDelay' in line:
        split_line = line.split('TDelay')
        time_delay = float(split_line[-1])
        if time_delay > 3:
            print line
```
