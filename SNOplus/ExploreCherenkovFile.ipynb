```python
import ROOT, rat
import os, sys
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
import rat
```

```python
infile_name = '/project/6001358/jpyanez/rat-scripts/cherenkov_mc/output2.root'
```

```python
try:
    ratreader.close()
    print 'Closing file before reopenning'
except:
    print 'No ratreader file. Opening it for the first time'
ratreader = rat.dsreader(infile_name)
```

```python
# Get one DS, run entry from the file
ds, run = ratreader.next()
```

```python
# Get the MC info for this DS entry (event)
mymc = ds.GetMC()
```

```python
# How many tracks inside (all particle segments)
print mymc.GetMCTrackCount()
```

```python
# Take one track inside and follow its steps
# The first will be the primary electron. There might be some secondaries produced.
itrack = 1
mytrack = mymc.GetMCTrack(itrack)
print 'PDG code:', mytrack.GetPDGCode()
```

```python
for istep in range(mytrack.GetMCTrackStepCount()):
    mystep = mytrack.GetMCTrackStep(istep)
    position = np.array(mystep.GetPosition())
    momentum = np.array(mystep.GetMomentum())
    end_volume = mystep.GetEndVolume()
    process = mystep.GetProcess()
    
    print 'Step ', istep
    print position, momentum
    print end_volume, process
```
