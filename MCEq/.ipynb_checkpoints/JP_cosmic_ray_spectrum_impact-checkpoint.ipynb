Simple muon and neutrino flux calculation
-----------------------------------------

This notebook demonstrates the basic use case of the solver. It calculates the atmospheric lepton flux, taking most of the configuration values from the 'global' config file `mceq_config.py`. The interaction model and the primary cosmic ray flux model are selected during initialization.

```python
#basic imports and ipython setup
%load_ext autoreload
%matplotlib inline
%autoreload 2
import os
import matplotlib.pyplot as plt
import numpy as np
os.chdir('/home/jp/projects/mceq/MCEq')

#import solver related modules
from MCEq.core import MCEqRun
from mceq_config import config
#import primary model choices
import CRFluxModels as pm
```

```python
config['debug_level']=0
fluxes = {}
mag = 3
```


# Differences between cosmic ray models

```python
# Create the model
model_tuples = ((pm.SimplePowerlaw27, '2.7'),
                (pm.PolyGonato, 'Polygonato'),
                (pm.HillasGaisser2012,"H4a"),
                (pm.HillasGaisser2012,"H3a"),
                (pm.GaisserStanevTilav, '4-gen'),
                (pm.GaisserHonda, 'GH'))
theta_angles_deg = np.rad2deg(np.arccos(np.linspace(0, 1, 16)))
```

```python
# I don't know if you always get 72 Energy points, but for now I'm taking them
mceq_run = MCEqRun(
    interaction_model='SIBYLL2.3_rc1',
    primary_model=model_tuples[0],
    theta_deg=0.,
    **config
    )
mceq_run.solve()
```

```python
epoints=len(mceq_run.e_grid)
e_grid = mceq_run.e_grid
mag=0
```

```python
for model_i in model_tuples:
    fluxes[model_i[1]] = {'mu_total':   np.zeros([epoints, len(theta_angles_deg)]),
                          'numu_total': np.zeros([epoints, len(theta_angles_deg)]),
                          'nue_total':  np.zeros([epoints, len(theta_angles_deg)])}
    mceq_run = MCEqRun(
        interaction_model='SIBYLL2.3_rc1',
        primary_model=model_i,
        theta_deg = 0.,
        **config
        )    
    for theta_i, one_theta_angle in enumerate(theta_angles_deg):

        mceq_run.set_theta_deg(one_theta_angle)
    
        # Solve
        mceq_run.solve()

        # Get results



        # total means conventional + prompt
        fluxes[model_i[1]]['mu_total'][:,theta_i] = (mceq_run.get_solution('total_mu+', mag)
                            + mceq_run.get_solution('total_mu-', mag))

        # same meaning of prefixes for muon neutrinos as for muons
        fluxes[model_i[1]]['numu_total'][:,theta_i] = (mceq_run.get_solution('total_numu', mag)
                              + mceq_run.get_solution('total_antinumu', mag))

        # same meaning of prefixes for electron neutrinos as for muons
        fluxes[model_i[1]]['nue_total'][:,theta_i] = (mceq_run.get_solution('total_nue', mag)
                             + mceq_run.get_solution('total_antinue', mag))

```

# Differences between interaction models

```python
import sys
import pickle
data = pickle.load(open('/home/jp/projects/MCEq/data/yield_dict.ppd'))
print 'Interaction models available'
preint_models = data.keys()


fluxes2 = {}
int_models = []
# Remove non-models
for one_model in preint_models:
    if not one_model in ['evec','ebins']:
        int_models.append(one_model)
        
int_models
```

```python
for had_model in int_models:
    
    fluxes2[had_model] = {}
    
    mceq_run = MCEqRun(
    interaction_model=had_model,
    primary_model=(pm.SimplePowerlaw27, '2.7'),
    theta_deg=80.0,   
    **config
    )

    # Solve
    mceq_run.solve()

    # Get results

    #obtain energy grid (nver changes) of the solution for the x-axis of the plots
    e_grid = mceq_run.e_grid

    # total means conventional + prompt
    fluxes2[had_model]['mu_total'] = (mceq_run.get_solution('total_mu+', mag)
                        + mceq_run.get_solution('total_mu-', mag))

    # same meaning of prefixes for muon neutrinos as for muons
    fluxes2[had_model]['numu_total'] = (mceq_run.get_solution('total_numu', mag)
                          + mceq_run.get_solution('total_antinumu', mag))

    # same meaning of prefixes for electron neutrinos as for muons
    fluxes2[had_model]['nue_total'] = (mceq_run.get_solution('total_nue', mag)
                         + mceq_run.get_solution('total_antinue', mag))

    # since there are no conventional tau neutrinos, prompt=total
    fluxes2[had_model]['nutau_total'] = (mceq_run.get_solution('total_nutau', mag)
                        + mceq_run.get_solution('total_antinutau', mag))
```

```python
#get path of the home directory + Desktop
save_pdf = False
desktop = os.path.join(os.path.expanduser("~"),'Desktop')
counter = 0
fig = []
for pref, lab in [('mu_',r'\mu'), ('numu_',r'\nu_\mu'), ('nue_',r'\nu_e')]: #, ('nutau_', r'\nu_\tau')]:
    
    fig.append(plt.figure(figsize=(7., 6)))
    
    ax1 = fig[-1].add_subplot(211)
    
    for model_i in int_models:
        plt.loglog(e_grid, fluxes2[model_i][pref + 'total'], ls='-', lw=1.5,
                  label=model_i)
    plt.ylim(1e-6,1)
    #plt.xlim(50,1e9)
    plt.legend(loc='upper right', ncol=2)


    ax2 = fig[-1].add_subplot(212, sharex=ax1)
    plt.ylabel(r"$\Phi_{" + lab + "}$ (E/GeV)$^{" + str(mag) +" }$" + 
               "(cm$^{2}$ s sr GeV)$^{-1}$")   
    
    for index in range(len(int_models)):
        plt.plot(e_grid, fluxes2[int_models[index]][pref + 'total'] / fluxes2[int_models[0]][pref + 'total'], 
                   ls='-', lw=1.5,
                    label=int_models[index])   
    plt.xlim([50, 1000])
    plt.ylim([0.7,1.3])

    plt.xlabel(r"$E_{{{0}}}$ [GeV]".format(lab))

    plt.tight_layout()
    if save_pdf: plt.savefig(os.path.join(desktop, pref + 'flux.pdf'))
```
