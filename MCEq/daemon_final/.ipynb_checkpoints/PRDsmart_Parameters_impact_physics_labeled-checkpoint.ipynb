```python
settings_file = '/home/jp/projects/atm_nu/ddm_results/final_settings/common_settings_splitVHE.pkl'
settings = pickle.load(open(settings_file, 'rb'))
```

```python
def get_vlist(target, plist):
    vlist = []
    for onep in plist:
        parameter = settings['zfactors_labeled'][onep]
        value = parameter[0]
        error = parameter[1]
        scale = target/(error/value)
        vlist.append(scale)
    return vlist
```

```python
def get_colors(plist):
    clist = []
    for onep in plist:
        clist.append(settings['hadron_colors1'][onep])
    return clist
```
