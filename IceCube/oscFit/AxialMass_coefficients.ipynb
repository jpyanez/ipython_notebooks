```python
import os, sys
import matplotlib.pyplot as plt
%matplotlib inline 
sys.path.append('/home/jp/projects/python_tools')
import jp_mpl as jplot
import pickle
import numpy as np
from copy import deepcopy
```

```python
def axialMassCoeffs(infile):
    data_dir = '/home/jp/projects/icecube/oscFit_data/tables'
    data = pickle.load(open(os.path.join(data_dir, infile)))
    xaxis = np.linspace(-0.5, 1.5, 101)
    resbool = data['scattering'] == 2
    # Resonant events
    c0, x = np.histogram(data['ma_res'][resbool,0],
                       xaxis)
    c1, x = np.histogram(data['ma_res'][resbool,1],
                        xaxis)
    plt.figure()
    jplot.unfilledBar(x, c0, color = 'blue', label='Quadratic')
    jplot.unfilledBar(x, c1, color = 'green', label='Linear')
    plt.legend(loc=0)

    # Resonant events
    resbool = data['scattering'] == 3
    c0, x = np.histogram(data['ma_qe'][resbool,0],
                       xaxis)
    c1, x = np.histogram(data['ma_qe'][resbool,1],
                        xaxis)
    plt.figure()
    jplot.unfilledBar(x, c0, color = 'blue', label='Quadratic')
    jplot.unfilledBar(x, c1, color = 'green', label='Linear')
    plt.legend(loc=0)
```

```python
infile   = 'sterile_selection_fixed/14550.pckl'
axialMassCoeffs(infile)
```
