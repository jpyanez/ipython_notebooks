```python
# Lets do it much easier. Do an interpolation and get the value from there
from scipy import interpolate

def getErrors(data):
    theta23_profile = 2*(data['llh'].min(axis=1) - data['llh'].min())
    theta23_values  = np.sin(data['var1_list'])**2

    yToFind = 1
    yreduced = np.array(theta23_profile) - yToFind
    freduced = interpolate.UnivariateSpline(theta23_values, 
                                        yreduced, s=0)
    
    theta23_error = np.mean(np.abs(np.sin(0.7)**2 - freduced.roots()))

    dm31_profile = 2*(data['llh'].min(axis=0) - data['llh'].min())
    dm31_values  = data['var2_list']
    yToFind = 1
    yreduced = np.array(dm31_profile) - yToFind
    freduced = interpolate.UnivariateSpline(dm31_values, 
                                        yreduced, s=0)
    dm31_error = np.mean(np.abs(0.0026 - freduced.roots()))
    
    return theta23_error, dm31_error
```

```python
mycolors = ['0.7','k','r','g','m','c','b']
contour_list = []
fit_names = ['DM31_Theta23_PRD','DM31_Theta23','FixLEXSections', 'FixNuFlux', 'FixGamma', 'FixMuons', 'FixDetector']
for i,scan_name in enumerate(fit_names):
    data = pickle.load(open(os.path.join(data_dir,'GenericNew_' + scan_name,'ScanResults.pckl')))
    contour_list.append(plt.contour(np.sin(data['var1_list'])**2, data['var2_list'],
                        2*(data['llh'].T-data['llh'].min()),
                        [delta_llh[0]], colors = (mycolors[i])))
    t23, dm31 = getErrors(data)
    print scan_name, t23, dm31
plt.legend([x.legend_elements()[0][0] for x in contour_list],
           fit_names,
           ncol = 3, loc = 0, numpoints = 2)

plt.xlabel(data['var1_name'], fontsize='large')
plt.ylabel(data['var2_name'], fontsize='large')
#plt.legend(loc=0, ncolumns=3)
```
