```python
loader =  dataLoader.dataLoader(bin_edges   = [10**np.linspace(0.75,2.15,11),
                                               np.arccos(np.linspace(sky_start,sky_stop,bins))[::-1],
                                                   np.array([pid_start,pid_stop])],
                                    observables = ['reco_energy', 'reco_zenith', 'delta_llh'],
                                    user = user,
                                    LEaxis = [],
                                    extra_cuts = {'bdt_score': (0.2,np.inf) },
                                    bkg_diff_is_syst = True,
                                    #user_init = [0],
                                    expected_years = 3,
                                    weight_keys = ['weight_e', 'weight_mu'],
                                    detsys_perflavor=False, # Calculate the detector variations per flavor
                                    legacy_detsys = False, # Calcualte the detector variations as in the PRD paper
                                    detailed_detsys=False, # Re-calculate the detector variations at each minimization step
                                    verbose = False)
```

```python
livetime = loader.sec2years*2.8
```

```python
h1 = loader.getNeutrinoHistograms()
h2 = loader.getNeutrinoHistograms({'w_pi':0.2})
```

```python
x= np.log10(loader.bin_edges[0])
y = np.cos(loader.bin_edges[1])
```

```python
def plotDiffs(baseline, modified, flavor):
    fig = plt.figure(figsize=(14,5))
    p1= fig.add_subplot(121)
    plt.title(flavor + '  A/B')
    plt.pcolor(x,y, 
               (modified[0][flavor+'_histo'][:,:,0]/baseline[0][flavor+'_histo'][:,:,0]).T)
    plt.xlabel('log10(Ereco/GeV)')
    plt.ylabel('cos(zenith_reco)')
    plt.colorbar()
    
    p2= fig.add_subplot(122)
    plt.title(flavor + '  (A-B)^2/A')
    plt.pcolor(x,y, 
               ((livetime*(modified[0][flavor+'_histo'][:,:,0]-baseline[0][flavor+'_histo'][:,:,0]))**2/
               (livetime*baseline[0][flavor+'_histo'][:,:,0])).T)
    plt.xlabel('log10(Ereco/GeV)')
    plt.ylabel('cos(zenith_reco)')
    plt.colorbar()
```

```python
plotDiffs(h1, h2, 'nue')
plotDiffs(h1, h2, 'numu')
plotDiffs(h1, h2, 'nutau')
```

```python
h1 = loader.getNeutrinoHistograms({'w_pi':0., 'w_k':0.})
h2 = loader.getNeutrinoHistograms({'w_pi':0., 'w_k':0.2})
```
