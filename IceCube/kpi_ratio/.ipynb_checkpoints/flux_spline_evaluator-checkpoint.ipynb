```python
import numpy as np
import cPickle as pickle
import matplotlib.pyplot as plt
```

Flux splines
------------

The spline files contain 3 dictionaries with the base flux and the two derivatives for pi-scaling and K-scaling. The energy range is limited to 1.XX GeV ro 10^7 GeV.

```python
spl_file = '/home/jpyanez/icecube/icecube-project/mceq_splines/spl_SIBYLL23c_H3a_demo.pkl'
```

```python
nominal, delta_wpi, delta_wk = pickle.load(open(spl_file))
```

```python
def get_flux(ltype, cos_theta, energy, w_pi, w_k, mag=0):
    """Returns flux in usual units 1/(GeV cm2 s sr)^-1.
    
    The ``mag`` parameter can be applied for plotting, scaling the reuslt with powers of energy.
    ltype can be one of 'mu+', 'mu-', 'numu', 'antinumu', 'nue', 'antinue'
    
    This thing is fastest for arrays of cos theta and energy, but single point access also works.
    """
    eval_single = lambda spl: (np.exp(spl(cos_theta, np.log(energy)))*energy**mag).T
    
    return (eval_single(nominal[ltype]) + 
            w_pi*eval_single(delta_wpi[ltype]) + 
            w_k*eval_single(delta_wk[ltype]))
```

```python
energy_grid = np.logspace(0.1,6,300)
cos_theta_grid = np.linspace(-1,0.,5)
```

```python
plt.loglog(energy_grid, 
           get_flux('nue',cos_theta_grid,energy_grid,0.,0.,mag=3) + 
           get_flux('antinue',cos_theta_grid,energy_grid,0.,0.,mag=3))

# Here w_pi = 0.30 and w_k = -0.30, scaling pions up by 30% and kaons down by -30% are applied
w_pi, w_k = 0.3, -0.35
plt.loglog(energy_grid, 
           get_flux('nue',cos_theta_grid,energy_grid,w_pi,w_k,mag=3) + 
           get_flux('antinue',cos_theta_grid,energy_grid,w_pi,w_k,mag=3),ls='--')
plt.ylim(1e-4,1e-1)
```
