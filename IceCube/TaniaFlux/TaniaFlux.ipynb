```python
sys.path.append('/home/jp/projects/icecube/tania_flux')
import TaniaFluxSp as TFlux
reload(TFlux)
```

```python
flux = TFlux.MCEqFluxSpline()
```

```python
energy = 100. # GeV
zenith = np.linspace(-1,1, 201)
mu_k = flux.EvaluateSpline('numu_from_k', energy, zenith)
mu_pi = flux.EvaluateSpline('numu_from_pion', energy, zenith)
mubar_k = flux.EvaluateSpline('antinum_from_k', energy, zenith)
mubar_pi = flux.EvaluateSpline('antinum_from_pion', energy, zenith)
```

```python
plt.plot(zenith, mu_pi, 'b', label = 'From Pions')
plt.plot(zenith, mubar_pi, '--b')
plt.plot(zenith, mu_k, 'g',label = 'From Kaons')
plt.plot(zenith, mubar_k, '--g')
plt.legend(loc=0)
plt.ylabel('Flux of muon (anti-)neutrinos [a.u.]')
plt.xlabel('cos(zenith)')
plt.title('Neutrino energy = '+str(energy)+' GeV')
plt.savefig('NuMu_split_' + str(energy) + '.png', dpi=200)
```
