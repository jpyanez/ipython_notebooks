```python
x = np.linspace(4, 80, 1000)
def mybias(energy):
    return np.exp(-energy/10.)/2 - 0.1
plt.plot(x,mybias(x))
```

```python
nevents = 100000
```

```python
res_i  = np.random.normal(0., 0.3, size = nevents)
energy = np.linspace(4., 80., nevents) 
```

```python
new_res = res_i + mybias(energy)
```

```python
plt.hist(new_res[:1000], bins=20)
```

```python
# Usual resolution plots
absres = np.abs(new_res)
```

```python
yaxis = np.linspace(0, 1.5, 31)
xaxis = np.linspace(4, 80, 15.)
bc, x, y = np.histogram2d(energy, absres, [xaxis, yaxis])
plt.pcolor(xaxis, yaxis, bc.T)
```

```python
median_resolution = np.zeros(xaxis.size-1)
for ebin in range(len(xaxis)-1):
    ebool = (energy > xaxis[ebin]) * (energy < xaxis[ebin+1])
    this_res = absres[ebool]
    median_resolution[ebin] = np.median(this_res)
```

```python
jplot.unfilledBar(xaxis, median_resolution)
```

```python
yaxis = np.linspace(-1.5, 1.5, 31)
xaxis = np.linspace(4, 80, 15.)
bc, x, y = np.histogram2d(energy, new_res, [xaxis, yaxis])
plt.pcolor(xaxis, yaxis, bc.T)
```

```python
resolution = np.zeros([2,xaxis.size-1])
for ebin in range(len(xaxis)-1):
    ebool = (energy > xaxis[ebin]) * (energy < xaxis[ebin+1])
    this_res = new_res[ebool]
    resolution[:,ebin] = [this_res.mean(), this_res.std()]
```

```python
yaxis = np.linspace(-1.5, 1.5, 31)
xaxis = np.linspace(4, 80, 15.)
bc, x, y = np.histogram2d(energy, new_res, [xaxis, yaxis])
plt.pcolor(xaxis, yaxis, bc.T, cmap='Blues')
newx = (xaxis[1:] + xaxis[:-1])/2.
plt.errorbar(newx, resolution[0, :], yerr = resolution[1,:], fmt='x', color = 'r')
plt.xlabel(r'$E_\nu\,\mathrm{(GeV)}$')
plt.ylabel(r'$\mathrm{Fractional\;energy\;resolution}$')
```

```python
resolution = np.zeros([3,xaxis.size-1])
for ebin in range(len(xaxis)-1):
    ebool = (energy > xaxis[ebin]) * (energy < xaxis[ebin+1])
    this_res = new_res[ebool]
    resolution[:,ebin] = np.concatenate(([np.mean(this_res)], np.percentile(this_res, [25, 75])))
    
fig= plt.figure(figsize=(7,5))
yaxis = np.linspace(-1.5, 1.5, 31)
xaxis = np.linspace(4, 80, 15.)
bc, x, y = np.histogram2d(energy, new_res, [xaxis, yaxis])
plt.pcolor(xaxis, yaxis, bc.T, cmap='Blues')
newx = (xaxis[1:] + xaxis[:-1])/2.
plt.plot(newx, resolution[0, :], 'xr')
plt.plot((newx, newx), resolution[1:,:], color = 'white')
plt.plot((newx+0.4, newx+0.4), resolution[1:,:], color = 'red')

#plt.errorbar(newx, np.zeros_like(newx), yerr = resolution[1:,:], color = 'red', fmt='')
#plt.axvline(newx, ymin = resolution[1,:], ymax = resolution[2,:], color = 'red')

print resolution[2,:] - resolution[1,:]

plt.xlabel(r'$E_\nu\,\mathrm{(GeV)}$')
plt.ylabel(r'$\mathrm{Fractional\;energy\;resolution}$')
fig.savefig('sarah_example.png', dpi=300)
```
